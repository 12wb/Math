# 求偏导数
# 求z =  x**2+3*x*y+y**2在点(1,2)处的偏导数

```python
from sympy import *
x,y = symbols('x y')
fn = x**2+3*x*y+y**2
fx = diff(fn,x)
fy = diff(fn,y) # 求偏导数
fx.subs({x:1,y:2}) # 求函数值  {x:1,y:2}即x=1 y=2
```

```python
fy.subs({x:1,y:2}).evalf(2)  # evalf为浮点数
```

# 求方向导数
# 求f(x,y,z) = xy+yz+zx在点(1,1,2)沿方向l的方向导数，其中l的方向分别为60°,45°，60°

```python
from sympy import * 
x,y,z = symbols('x y z')
fn = x*y+y*z+z*x
fx = diff(fn,x) # 先求偏导数
fy = diff(fn,y)
fz = diff(fn,z)
fp = fx*cos(pi/3)+fy*cos(pi/4)+fz*cos(pi/3)  # 方向余弦*每一项偏导数 pi为π
print(fp)
fp.subs({x:1,y:1,z:2}).evalf()
```

# 求梯度
# 设f(x,y)=1/2(x**2+y**2),p0(1,1),求
# (1)f(x,y)在p0处增加最快的方向以及f(x,y)沿这个方向的方向导数
# (2)f(x,y)在p0处减少最快的方向以及f(x,y)沿这个方向的方向导数
# (3)f(x,y)在p0处的变化率为零的方向

```python
from sympy.vector import * # 这是一个向量工具库
from sympy.vector.coordsysrect import CoordSys3D
C = CoordSys3D('C') # 选取坐标轴 
f = 1/2*(C.x**2+C.y**2)
gradient(f)
```

```python
res = gradient(f) # 求梯度
res.subs({C.x:1,C.y:1})
```

```python
res.normalize() # 求向量方向
```

```python
res.magnitude() # 求向量的模，梯度的模即最大的方向导数
```
