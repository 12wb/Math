# 作业三

```python
from sympy import *
x,y,z = symbols('x y z')
fn = x*y+y*z+z*x
fx = diff(fn,x)
fy = diff(fn,y)
fz = diff(fn,z)
fp = fx*cos(pi/3)+fy*cos(pi/4)+fz*cos(pi/3) # 方向余弦*每一项偏导数   pi为π
```

```python
fp
```

```python
fp.subs({x:1,y:1,z:2}).evalf()
```
