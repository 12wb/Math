# 求函数f(x) = e**x的n阶麦克劳林展开式
# n =10
# (画出变化率)

```python
from sympy import * 
```

```python
x = symbols('x')
Fn = E**x #目标函数
fn = series(Fn,x,n=10)  # 泰勒展开式
fn
```

```python
plotting.plot(fn.subs(O(x**10),0),(x,0,10))  # 麦克劳林展开式   O代表Rn**(X)
plotting.plot(Fn,(x,0,10)) # 原函数
```
